```python
import numpy as np
import os

base_dir = '/home/gui/snippets_test/cover_area/default_res'

for folder in os.listdir(base_dir):
    folder_path = os.path.join(base_dir, folder)
    
    for file in os.listdir(folder_path):
        if file.endswith('.npy'):
            file_path = os.path.join(folder_path, file)
            array = np.load(file_path)

            # Certifique-se de que o array tem pelo menos 3 snippets
            if len(array) < 3:
                print("O array nÃ£o tem 3 snippets.")
            else:
                # Salvar cada snippet em um novo arquivo .npy
                for i in range(3):
                    snippet_filename = f"snippet_hr_{i}.npy" if 'hr' in file else f"snippet_motion_{i}.npy"
                    save_path = os.path.join(folder_path, snippet_filename)
                    np.save(save_path, array[i])
                    print(f"Snippet {i} salvo em {save_path}")
```

```python
import numpy as np
import os
from dtaidistance import dtw

dtw_results = {}

base_path = '/home/gui/snippets_test/cover_area'

for folder in os.listdir(os.path.join(base_path, 'default_res')):
    folder_path = os.path.join(base_path, 'default_res', folder)
    dtw_results[folder] = {'hr': {}, 'motion': {}}
    for file in os.listdir(folder_path):
        if file.endswith('.npy') and 'snippet' in file:
            snippet_default = np.load(os.path.join(folder_path, file))
            snippet_erb = np.load(os.path.join(base_path, 'erb_res', folder, file))
            
            distance = dtw.distance(snippet_default, snippet_erb)
            if 'hr' in file:
                dtw_results[folder]['hr'][file] = distance
            else:
                dtw_results[folder]['motion'][file] = distance
```

```python
import json

with open('/home/gui/snippets_test/cover_area/dtw_results.json', 'w') as f:
    json.dump(dtw_results, f, indent=4)
```

```python
import plotly.graph_objects as go

fig = go.Figure()

for file, values in dtw_results.items():
    for category, snippets in values.items():
        for snippet, value in snippets.items():
            fig.add_trace(go.Bar(
                x=[file],
                y=[value],
                name=f"{category} - {snippet}",
                hoverinfo="y+name",
                marker=dict(line=dict(width=0.8))
            ))

# Configurando layout
fig.update_layout(
    title="VisualizaÃ§Ã£o de Dados de 'hr' e 'motion'",
    xaxis=dict(title="Arquivo (Data e Hora)"),
    yaxis=dict(title="Valor"),
    barmode="group",
    legend_title="Categoria e Snippet",
    hovermode="x"
)

# Exibir grÃ¡fico
fig.show()
```
